```python
video_ids = []

with open("video_ids.txt", "r", encoding="utf-8") as f:
    for line in f:
        vid = line.strip()
        if vid:   # skip blank lines
            video_ids.append(vid)

print("Extracted video IDs:", video_ids)
print("Count:", len(video_ids))
```

```python
import os
import re
import requests
from youtube_transcript_api._api import YouTubeTranscriptApi

os.makedirs("transcripts", exist_ok=True)
yt_api = YouTubeTranscriptApi()

for video_id in video_ids:
    try:
        print(f"\nProcessing video: {video_id}")

        # Get transcript
        transcript_list = yt_api.list(video_id)
        transcript = transcript_list.find_transcript(['en'])
        transcript_data = transcript.fetch()
        text = " ".join(t.text for t in transcript_data)

        # Get title from oEmbed
        oembed_url = f"https://www.youtube.com/oembed?url=https://www.youtube.com/watch?v={video_id}&format=json"
        title = requests.get(oembed_url).json()['title']

        # Create safe filename
        safe_title = re.sub(r"[^a-zA-Z0-9_\- ]", "", title).replace(" ", "_")
        filename = f"transcripts/{safe_title}.txt"

        # Save transcript
        with open(filename, "w", encoding="utf-8") as f:
            f.write(text)

        print(f"Saved: {filename}")

    except Exception as e:
        print(f"Error for {video_id}: {e}")
```

```python
def chunk_text(text, chunk_size=500, overlap=100):
    """
    Splits text into overlapping chunks.
    chunk_size = number of characters per chunk
    overlap = number of characters that overlap between chunks
    """
    chunks = []
    start = 0
    text_length = len(text)

    while start < text_length:
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += (chunk_size - overlap)  # move forward by chunk_size - overlap

    return chunks
```

```python
import os
all_chunks = []  # this will store dictionaries of: {"video", "chunk", "text"}

transcript_folder = "transcripts"

for filename in os.listdir(transcript_folder):
    if filename.endswith(".txt"):
        filepath = os.path.join(transcript_folder, filename)

        with open(filepath, "r", encoding="utf-8") as f:
            text = f.read()

        chunks = chunk_text(text)

        for i, chunk in enumerate(chunks):
            all_chunks.append({
                "video": filename,
                "chunk_id": i,
                "text": chunk
            })

print("Number of chunks:", len(all_chunks))
print("Example chunk:", all_chunks[0])
```

```python
import json

with open("chunks.json", "w", encoding="utf-8") as f:
    json.dump(all_chunks, f, indent=2, ensure_ascii=False)

print("Saved chunks.json with", len(all_chunks), "chunks.")
```
